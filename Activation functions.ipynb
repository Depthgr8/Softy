# Neural Activation Functions

1. Basic neural network calculations
2. Activation function implementations
3. Visualization of different activation functions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# First example
X = np.array([[0.2], [0.4], [0.6]])
W = np.array([[0.1], [0.2], [0.3]])
Y = W.T @ X  # Matrix multiplication
print("First output value:", Y[0,0])

# Second example
X = np.array([[1], [0]])
W = np.array([[2], [-1]])
Y = W.T @ X
print("\nSecond output value:", Y[0,0])
```

## Activation Function visualization configuration

```python
# Variable initialization
xmin, xmax = -3, 3
dp = 100
sigma = 2
x = np.linspace(xmin, xmax, dp)
```

## Binary Step Function

```python
def binary_step(x, threshold=0):
    return (x >= threshold).astype(np.float64)

y1 = binary_step(x, 0)
plt.figure(figsize=(5, 2))
plt.plot(x, y1)
plt.title('Binary Step Activation')
plt.grid(True)
plt.show()
```

## Binary Sigmoid

```python
y2 = 1 / (1 + np.exp(-sigma * x))
plt.figure(figsize=(5, 2))
plt.plot(x, y2)
plt.title('Binary Sigmoid Activation')
plt.grid(True)
plt.show()
```

## Bipolar Sigmoid

```python
y3 = (2 * y2) - 1
plt.figure(figsize=(5, 2))
plt.plot(x, y3)
plt.title('Bipolar Sigmoid Activation')
plt.grid(True)
plt.show()
```

## Hyperbolic Tangent

```python
y4 = np.tanh(x)
plt.figure(figsize=(5, 2))
plt.plot(x, y4)
plt.title('Hyperbolic Tangent Activation')
plt.grid(True)
plt.show()
```
