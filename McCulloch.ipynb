# Neural network architecture
## McCulloch-Pitts Neuron model

#### Inspired by Biological neural network

*   Simplest model
*   Either fires or doesn't fire (Binary Input/Output)
*   Positive weights indicate excitatory connections  (promotes firing)
*   Negative weights represent inhibitory connections (suppresses firing)
*   Takes a discrete time step for firing. Output neuron fires at t+1 time put when Input neuron fires at t time point.



```python
import tensorflow as tf
def mcculloh_pits_neuron(inputs,weights,threshold):
    mul = tf.multiply(inputs,weights)
    weighted_sum = tf.reduce_sum(mul)

    output = tf.cond(weighted_sum >= threshold,lambda:1.0,lambda:0)
    print("Y_in = ",weighted_sum,"; Y = " output)
    return output

inputs = tf.constant([0.5,0.3,0.8],dtype=tf.float32)
weights = tf.constant([0.2,0.4,0.6],dtype=tf.float32)
threshold = tf.constant(0.6,dtype=tf.float32)
output = mcculloh_pits_neuron(inputs,weights,threshold)
print("output",output)
```
